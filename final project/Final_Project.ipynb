## Contig Length Histogram

```python
# Plotting Contig Length

from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import generic_protein
from Bio import SeqIO
from Bio import Phylo
import numpy as np
import matplotlib.pyplot as plt
import time
import re

contigs = []
contig_lens = []
for seq_rec in SeqIO.parse("/bigdata/FinalProject_groups/Group_14/contigs.fasta", "fasta"):   
    contigs.append(seq_rec.seq)
    contig_lens.append(len(str(seq_rec.seq)))
    

# Stolen from: https://realpython.com/python-histograms/

import matplotlib.pyplot as plt

# An "interface" to matplotlib.axes.Axes.hist() method
n, bins, patches = plt.hist(x=contig_lens, bins='auto',
                            alpha=0.7, rwidth=0.85)
# plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Contig Lengths')
```

```python
# Excluding values of 200 000 or higher:
# Excluding valaues at 150 and 156:
# Stolen from: https://realpython.com/python-histograms/

cov_new = [c for c in contig_lens if c < 200000]
import matplotlib.pyplot as plt

# An "interface" to matplotlib.axes.Axes.hist() method
n, bins, patches = plt.hist(x=cov_new, bins='auto',
                            alpha=0.7, rwidth=0.85)
# plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Contig Lengths')
```
