### Simulating ON/OFF state of the Riboswitches

```python
yes1_cnstr = ""
for i in np.arange(len(yes_1)):
    if (i < 26) | (i > 47):
        yes1_cnstr = yes1_cnstr + "."
    else:
        yes1_cnstr = yes1_cnstr + "x"
yes1_cnstr
```

```python
f = open("YES-1_contraints.txt","w+")
f.write(yes_1 + "\n")
f.write(yes1_cnstr)
f.close()
```

```python
!RNAfold -C < YES-1_contraints.txt
```

```python
!mv rna.ps yes_1_cons.pdf
img_yes1_cons = Image(filename="yes_1_cons.pdf")
img_yes1_cons
```
